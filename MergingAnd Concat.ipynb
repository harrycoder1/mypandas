# Merge

```python
import pandas as pd  
d1 = pd.DataFrame({"A":[1,2,3,4] ,"B" :[11,22,33,44]})
d2 = pd.DataFrame({"A":[1,2,3,4] ,"C" :[21,32,43,34]})
pd.merge(d1,d2 , on="A")

```

```python
import pandas as pd  
d1 = pd.DataFrame({"A":[1,2,3,4] ,"B" :[11,22,33,44]})
d2 = pd.DataFrame({"A":[1,2,3,5] ,"C" :[21,32,43,34]})
pd.merge(d1,d2 , on="A") # only common data will be merged
```

```python
pd.merge(d1 , d2 , how="inner")
```

```python
pd.merge(d1 , d2 , how="left")
```

```python
pd.merge(d1 , d2 , how="right")
```

```python
pd.merge(d1 , d2 , how="outer")
```

```python
pd.merge(d1 , d2 ,indicator=True ,how="outer" )
```

```python
import pandas as pd  
d1 = pd.DataFrame({"A":[1,2,3,4] ,"B" :[11,22,33,44]})
d2 = pd.DataFrame({"A":[1,2,3,5] ,"B" :[21,32,43,34]})
pd.merge(d1,d2 ,left_index=True ,right_index=True  )
```
