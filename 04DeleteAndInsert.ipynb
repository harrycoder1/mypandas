#  Insert in Pandas

```python
import pandas as pd
var =pd.DataFrame({"A":[1,2,3,4,5] ,"B":[1,2,3,4,5]})

var
```

```python
var.insert(1, "python1" , var["A"]**2)
var
```

```python
var.insert(2,"pythe2" , [11,12,13,14,15])
var
```

```python
var["harish"] = var["A"][1:4]
var
```
