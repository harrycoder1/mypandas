# Replace

```python
import pandas as pd
csv = pd.read_csv("E:\\learing code\\Data Science\\pandas\\record.csv" )
csv
```

```python
csv.replace(to_replace="Omkar Eknath Vaidya " ,value="kishor")
```

```python
csv.replace([x for x in range(1,35)] , 22)
```

```python
csv.replace({"Name":"[A-Z]"} , 22 , regex= True)
```

```python
csv.replace("3rd year",method= "ffill")
```

```python
csv.replace("2nd year" , method="ffill",limit=2)
```

```python
csv.replace("2nd year" , method="ffill",limit=2 ,inplace=True)
csv
```
